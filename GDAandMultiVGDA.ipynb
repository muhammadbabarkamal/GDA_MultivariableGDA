## Q- 1. The median price (in thousands of dollars) of existing homes in a certain metropolitan area from 1989 to 1999 was:


USD 86400,89800,92800,96000,99600,103100,106300,109500,113300,120000,129500

Years 1989,1990,1991,1992,1993,1994,1995,1996,1997,1998,1999 


* Find an equation of the least squares line 
for these data by finding optimal parameter
through Gradient Descent Algorithm. Initially take parameters as zero and learning
parameter 0.01.
* Estimate the median price of a house in the year 2010, 2020 and the year 2025, assuming
that the trend continues.




```python
import numpy as np
label = np.array([1989,1990,1991,1992,1993,1994,1995,1996,1997,1998,1999])
# normalise the values by min-max normalisation
normalised_value = np.array([((86.4-86.4)/(129.5-86.4)),((89.8-86.4)/(129.5-86.4)),((92.8-86.4)/(129.5-86.4)),((96.0-86.4)/(129.5-86.4)),((99.6-86.4)/(129.5-86.4)),((103.1-86.4)/(129.5-86.4)),((106.3-86.4)/(129.5-86.4)),((109.5-86.4)/(129.5-86.4)),((113.3-86.4)/(129.5-86.4)),((120.0-86.4)/(129.5-86.4)),((129.5-86.4)/(129.5-86.4))])

def GDA(value,label):
    norm =129.5
    n1=26000
    iterations = 156001
    m = len(value)
    lr = 0.001
    ϴ0=0
    ϴ1=0
    for i in range(iterations):
        ŷ = ϴ0 + ϴ1 * value
        cost = (1/20) * sum([val**2 for val in (ŷ)-label])
        Δϴ0 = (1/m)*sum(ŷ-label)
        Δϴ1 = (1/m)*sum((ŷ-label)*value)
        ϴ0 = ϴ0 - lr * Δϴ0
        ϴ1 = ϴ1 - lr * Δϴ1
        if i % n1 == 0:
          print ("ϴ0 {}, ϴ1 {}, cost {} iteration {}".format(ϴ0,ϴ1,cost, i))
        predict2010=((2010-ϴ0)/ϴ1)*norm
        predict2020=((2020-ϴ0)/ϴ1)*norm
        predict2025=((2025-ϴ0)/ϴ1)*norm
        predict2035=((2035-ϴ0)/ϴ1)*norm
    return (predict2010,predict2020,predict2025,predict2035)

predict2010,predict2020,predict2025,predict2035=GDA(normalised_value,label)
predict2010=int(predict2010);predict2020=int(predict2020);predict2025=int(predict2025);predict2035=int(predict2035)
print("prediction for \n Price for year 2010: {},000 USD \n Price for year 2020: {},000 USD \n Price for year 2025: {},000 USD \n Price for year 2035: {},000 USD".format(predict2010,predict2020,predict2025,predict2035))
```
